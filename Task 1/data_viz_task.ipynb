```python
# Import libraries 
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns

# Inline indicates graphs should be presented as a cell output
%matplotlib inline
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
```

### 1. Boxplot

```python
# Boxplot for Revs per mile for selected manufacturers
# Filter data for selected manufacturers

# Filter data for selected manufacturers
selected_manufacturers = car_data[
    car_data['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])
]

# Calculate the average revs per mile for each manufacturer
avg_revs_per_mile = selected_manufacturers.groupby(
    'Manufacturer'
)['Rev.per.mile'].mean().reset_index()

# Create the boxplot
plt.figure(figsize=(7, 6))
sns.boxplot(
    x='Manufacturer',
    y='Rev.per.mile',
    data=selected_manufacturers
)
plt.title('Revs per Mile for Selected Manufacturers')
plt.show()

```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
The visualisation displays that the Suzuki manufacturer has limited data,
but still has significantly higher Revs per mile than other 
manufacturers at a median of nearly 3400.
Hyundai and Toyota have more variability in data than Suzuki but Hyundai 
has the highest revs per mile with a median near 2700 if the limited data of 
Suzuki is taken into account and the outlier of Toyota at approximately 3500.

### 2. Histogram

```python

# Melt the DataFrame to create a long format for easy plotting
car_data_melted = car_data.melt(id_vars='Type', value_vars=['MPG.highway',
                                                            'MPG.city'], 
                                var_name='MPG_Type', value_name='MPG')

# Plot the histogram with bars side by side for highway and city MPG
plt.figure(figsize=(10, 6))
sns.histplot(data=car_data_melted, x='MPG', hue='MPG_Type', 
             multiple='dodge', bins=10)

# Labels 
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Frequency')
plt.title('Distribution of MPG for Highway and City by Car')

# Show plot
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
The histogram compares the distribution of miles per 
gallon (MPG) for city driving (MPG city) and highway driving (MPG highway).
X-axis (Miles Per Gallon): This represents the MPG, which indicates fuel efficiency.
Y-axis (Frequency): This represents the number of cars that achieve a particular MPG.

Orange Bars (MPG.city): These show the frequency distribution of MPG for city driving.
Blue Bars (MPG.highway): These show the frequency distribution of MPG for highway driving

Higher Frequency at Lower MPG: The orange bars (city driving) are more frequent at lower MPG values (between 15 to 25 MPG), indicating that most cars get lower MPG in the city.
Higher Frequency at Higher MPG: The blue bars (highway driving) are more frequent at higher MPG values (around 25 to 35 MPG and beyond), indicating that most cars are more fuel-efficient on the highway.

Fuel Efficiency: The histogram shows that cars generally achieve better MPG on the highway compared to city driving. Therefore, it is generally more fuel-efficient to drive on the highway.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
# Create the line plot
plt.figure(figsize=(7, 5))
sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_data)

# Title of line plot 
plt.title('Relationship Between Wheelbase and Turn circle')

# Show plot
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
Based on the trend of the graph, as the wheelbase increases the turn circle
increases as well. Therefore the graph shows a positive correlation between the 
wheelbase and turn circle and some variablility.


### 4. Barplot

```python
# Barplot for mean horsepower
# Create the bar plot

plt.figure(figsize=(7, 6))
sns.barplot(
    x='Type',
    y='Horsepower',
    data=car_data, # Get data from the file
    estimator=lambda x: x.mean(),
    # Calculate mean values of the horsepower
    errorbar=None
)

# Labels and x-axis ticks
plt.title('Mean Horsepower by Car Type')
plt.xticks(rotation=45)
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
According to the barplot small cars have the lowest mean horse power and 
large cars have the highest mean horsepower. This shows that the horsepower 
will generally increase based on the size of the car.
